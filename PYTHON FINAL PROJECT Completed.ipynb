🧹 Handling Missing Values

```python
df.isnull().sum()

# Fill text columns
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].fillna('Not Reported')

# Fill numeric columns
for col in df.select_dtypes(include=['int64', 'float64']).columns:
    df[col] = df[col].fillna(0)

df.isnull().sum()
```

```python
# Check total missing values in each column after imputation
df.isnull().sum()
df.isnull().sum().sum()
```

<h2 style="color:#b57edc; font-weight:700;">
♻️ Removing Duplicates
</h2>

```python
#Check the number of duplicate rows
duplicate_count = df.duplicated().sum()
print("Number of duplicate rows:", duplicate_count)
```

```python
# Remove duplicate rows

df = df.drop_duplicates()
```

```python
# Verify duplicates removed

print("Number of duplicate rows after removal:", df.duplicated().sum())
```

<h2 style="color:#008080; font-weight:700;">
🔧 Correcting Data Types
</h2>

```python
# Convert 'date_added' to datetime

if 'date_added' in df.columns:
    df['date_added'] = pd.to_datetime(df['date_added'])
```

```python
# Check data types after conversion

df.dtypes
```

<h2 style="color:#8E7AB5;font-weight:700;">✨ Creating Derived Columns</h2>

```python
# Create 'side_effect_count' from 'side_effects' column
if 'side_effects' in df.columns:
    df['side_effect_count'] = df['side_effects'].str.split(',').str.len()
```

```python
# Create 'adverse_effect_flag'
if 'side_effects' in df.columns:
    df['adverse_effect_flag'] = df['side_effects'].apply(
        lambda x: 'Yes' if pd.notna(x) else 'No'
    )
```

```python
# Create 'safety_level' based on side effect count
def safety_level(count):
    if count == 0 or pd.isna(count):
        return 'Low Risk'
    elif count <= 3:
        return 'Moderate Risk'
    else:
        return 'High Risk'

df['safety_level'] = df['side_effect_count'].apply(safety_level)
```

```python
# Create 'severity_score'
severity_map = {
    'Low Risk': 1,
    'Moderate Risk': 2,
    'High Risk': 3
}

df['severity_score'] = df['safety_level'].map(severity_map)
```

```python
# Create 'drug_name_length'
if 'drug_name' in df.columns:
    df['drug_name_length'] = df['drug_name'].str.len()
```

```python
# Create 'usage_risk_index'
df['usage_risk_index'] = df['side_effect_count'] * df['severity_score']
```

```python
df[['side_effect_count', 'adverse_effect_flag', 'safety_level',
    'severity_score', 'usage_risk_index']].head()
```

Derived columns were created to enhance drug safety analysis by quantifying
side effects, categorizing risk levels, and enabling better comparisons
between drugs.


<h2 style="color:#E50914;font-weight:700;">📊 Filtering Data </h2>

```python
# Filter drugs with high safety risk
high_risk_drugs = df[df['safety_level'] == 'High Risk']

high_risk_drugs.head()
```

```python
# Filter drugs with reported side effects
drugs_with_side_effects = df[df['adverse_effect_flag'] == 'Yes']

drugs_with_side_effects.head()
```

```python
# Filter drugs with more than 3 side effects
high_side_effect_count = df[df['side_effect_count'] > 3]

high_side_effect_count.head()
```

```python
# Filter low risk drugs
low_risk_drugs = df[df['safety_level'] == 'Low Risk']

low_risk_drugs.head()
```

```python
# Drugs with side effects AND high risk
high_risk_with_effects = df[
    (df['adverse_effect_flag'] == 'Yes') &
    (df['safety_level'] == 'High Risk')
]

high_risk_with_effects.head()
```

```python
# Top 10 drugs with highest usage risk index
top_risky_drugs = df.sort_values(
    by='usage_risk_index',
    ascending=False
).head(10)

top_risky_drugs
```

Filtering techniques were applied to isolate high-risk drugs, drugs with
reported side effects, and low-risk drugs to support focused safety analysis.


<h2 style="color:#8E7AB5;font-weight:700;">📊 Aggregating Data</h2>

```python
# Count number of drugs in each safety level
safety_level_count = df.groupby('safety_level').size().reset_index(name='drug_count')

safety_level_count
```

```python
# Average side effect count per safety level
avg_side_effects = df.groupby('safety_level')['side_effect_count'].mean().reset_index()

avg_side_effects
```

```python
# Total side effects per drug
drug_side_effects = df.groupby('drug_name')['side_effect_count'].sum().reset_index()

drug_side_effects.sort_values(by='side_effect_count', ascending=False).head(10)
```

```python
# Count of high-risk drugs
high_risk_count = df[df['safety_level'] == 'High Risk'].shape[0]

high_risk_count
```

```python
# Average usage risk index per safety level
risk_index_avg = df.groupby('safety_level')['usage_risk_index'].mean().reset_index()

risk_index_avg
```

```python
# Count drugs by adverse effect flag
effect_flag_count = df.groupby('adverse_effect_flag').size().reset_index(name='count')

effect_flag_count
```

Aggregation techniques were applied to summarize drug safety data by grouping
records based on safety level, side effect count, and risk indicators to
derive meaningful insights.


### Data Pre-processing Summary

All missing values were handled appropriately using fillna().
Duplicate records were removed.
Data types were verified and corrected where necessary.
Derived columns were created to enhance analysis.
The dataset is now clean and ready for exploratory data analysis.

### 1.📊 DRUG CLASS DISTRIBUTION (Bar Chart)

```python
import matplotlib.pyplot as plt

# Get Top 10 drug classes
top10 = df['drug_classes'].value_counts().head(10)

plt.figure(figsize=(10,6))

colors = ['#FF6F91', '#845EC2', '#00C9A7', '#FFC75F', '#4D96FF',
          '#FF9671', '#00C2A8', '#C34A36', '#0081CF', '#F9F871']

plt.bar(
    top10.index,
    top10.values,
    color=colors
)

plt.title("TOP 10 DRUG CLASSES BY COUNT",
          fontsize=14, fontweight='bold')

plt.xlabel("DRUG CLASS")
plt.ylabel("COUNT")

plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.3)

plt.tight_layout()
plt.show()
```

**Insight:**  
The bar chart shows that certain drug classes dominate the dataset.
The highest represented class contains significantly more drugs
compared to other categories. This indicates concentration of
drug development within specific therapeutic areas.


### 2.📊 SAFETY LEVEL DISTRIBUTION USING COUNT PLOT


```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

sns.countplot(
    data=df,
    x='safety_level',
    palette='Set2'
)

plt.title("Safety Level Distribution", fontsize=16, fontweight='bold')
plt.xlabel("Safety Level", fontsize=12)
plt.ylabel("Count", fontsize=12)

sns.despine()
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

**Insight:**  
The count plot shows the number of drugs in each safety category. 
It helps identify which safety level is most common in the dataset 
and highlights the overall distribution pattern of drug risk levels.


### 3.📊 PROPORTION OF DRUGS BY SAFETY LEVEL (Donut Chart)



```python
import matplotlib.pyplot as plt

plt.figure(figsize=(7,7))

counts = df['safety_level'].value_counts()

colors = ['#6A5ACD', '#20B2AA', '#FF7F50']  # Attractive color palette

plt.pie(
    counts,
    labels=counts.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=colors,
    wedgeprops={'edgecolor': 'white'}
)

# Create white circle in center (Donut effect)
centre_circle = plt.Circle((0,0), 0.65, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title("Safety Level Distribution", fontsize=16, fontweight='bold')

plt.tight_layout()
plt.show()
```

**Insight:**  
The donut chart shows the proportion of drugs across different safety levels. 
The majority category indicates the dominant safety classification, while 
smaller segments highlight drugs requiring higher caution.


### 4. 📊SEVERITY vs SIDE EFFECT COUNT (Scatter Plot)

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

plt.scatter(
    df['side_effect_count'],
    df['severity_score'],
    alpha=0.6,
    edgecolors='black'
)

plt.title("Severity Score vs Side Effect Count", fontsize=16, fontweight='bold')
plt.xlabel("Side Effect Count", fontsize=12)
plt.ylabel("Severity Score", fontsize=12)

plt.grid(True, linestyle='--', alpha=0.5)

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```

**Insight:**  
The scatter plot illustrates the relationship between side effect count 
and severity score. A positive trend can be observed, indicating that 
drugs with a higher number of reported side effects tend to have higher 
severity scores. However, some points deviate from the general pattern, 
suggesting variability in drug response and risk levels.


### 5. 📊 SIDE EFFECT COUNT ACROSS SAFETY LEVELS (Box Plot)


```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

sns.boxplot(
    data=df,
    x='safety_level',
    y='side_effect_count',
    palette='Pastel1'
)

plt.title("Side Effect Count Across Safety Levels", fontsize=16, fontweight='bold')
plt.xlabel("Safety Level", fontsize=12)
plt.ylabel("Side Effect Count", fontsize=12)

sns.despine()
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

**Insight:**  
The boxplots show that higher safety risk categories generally have higher 
median severity scores and side effect counts. The presence of outliers 
indicates that some drugs exhibit extreme values, which may require 
additional monitoring or regulatory attention.


### 6.📊 CORRELATION ANALYSIS OF NUMERICAL VARIABLES (Heat Map)


```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

correlation = df[['side_effect_count', 'severity_score']].corr()

sns.heatmap(
    correlation,
    annot=True,
    cmap='coolwarm',
    linewidths=0.5
)

plt.title("Correlation Heatmap", fontsize=16, fontweight='bold')

plt.tight_layout()
plt.show()
```

**Insight:**  
The heatmap shows the correlation between numerical variables. 
A positive correlation between side effect count and severity score 
suggests that drugs with more side effects tend to have higher severity levels.


### 7.📊 AVERAGE SEVERITY SCORE BY SAFETY LEVEL (Horizontal Bar Chart)


```python
import matplotlib.pyplot as plt

# Calculate average severity per safety level
avg_severity = df.groupby('safety_level')['severity_score'].mean().sort_values()

plt.figure(figsize=(8,6))

colors = ['#FF6F91', '#845EC2']  # Pink & Purple (adjust if more levels)

plt.barh(
    avg_severity.index,
    avg_severity.values,
    color=colors[:len(avg_severity)]
)

plt.title("AVERAGE SEVERITY SCORE BY SAFETY LEVEL",
          fontsize=14, fontweight='bold')

plt.xlabel("AVERAGE SEVERITY SCORE")
plt.ylabel("SAFETY LEVEL")

plt.grid(axis='x', linestyle='--', alpha=0.4)

plt.tight_layout()
plt.show()
```

**Insight:**  
The horizontal bar chart shows the average severity score across different safety levels. It is observed that the High Risk category has a higher average severity score compared to the Moderate Risk category. This indicates that drugs classified as High Risk tend to be associated with more severe side effects. The difference between the categories highlights the effectiveness of the safety classification system.


### 8.📊 SEVERITY SCORE TREND BY SAFETY LEVEL

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate average severity score per safety level
avg_severity = df.groupby('safety_level')['severity_score'].mean().reset_index()

plt.figure(figsize=(8,6))

sns.lineplot(
    data=avg_severity,
    x='safety_level',
    y='severity_score',
    marker='o'
)

plt.title("SEVERITY SCORE TREND BY SAFETY LEVEL", fontsize=14, fontweight='bold')
plt.xlabel("Safety Level")
plt.ylabel("Average Severity Score")

plt.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

**Insight:**  
The line plot illustrates the trend of severity score in relation to side effect count across different safety levels. It shows that as the number of side effects increases, the severity score also tends to increase. Additionally, the High Risk category consistently demonstrates higher severity scores compared to the Moderate Risk category, indicating a stronger impact of side effects in high-risk drugs.


### 9.📊 DISTRIBUTION OF SIDE EFFECT COUNT (Histogram + Density Curve)

```python
plt.figure(figsize=(9,6))

sns.histplot(
    df['side_effect_count'],
    bins=15,
    kde=True,
    color='#00C9A7',
    edgecolor='black'
)

plt.title("DISTRIBUTION OF SIDE EFFECT COUNT",
          fontsize=16, fontweight='bold')

plt.xlabel("SIDE EFFECT COUNT", fontsize=12)
plt.ylabel("FREQUENCY", fontsize=12)

plt.tight_layout()
plt.show()
```

**Insight:**  
The histogram displays the frequency distribution of side effect counts in the dataset. Most drugs cluster around a specific range of side effect counts, while fewer drugs exhibit extremely high counts. The smooth density curve highlights the overall distribution pattern, helping identify whether the data is concentrated or widely spread.


### 10.📊 INDIVIDUAL SEVERITY SCORE DISTRIBUTION BY SAFETY LEVEL (Strip Plot)

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")

plt.figure(figsize=(9,6))

sns.stripplot(
    data=df,
    x='safety_level',
    y='severity_score',
    palette=['#FF6F91', '#845EC2'],
    jitter=True,
    size=6,
    alpha=0.7
)

plt.title("INDIVIDUAL SEVERITY SCORE DISTRIBUTION BY SAFETY LEVEL",
          fontsize=16, fontweight='bold')

plt.xlabel("SAFETY LEVEL", fontsize=12)
plt.ylabel("SEVERITY SCORE", fontsize=12)

plt.tight_layout()
plt.show()
```

**Insight:**  
The strip plot displays individual severity score data points across different safety levels. It reveals that the High Risk category generally has higher severity values compared to the Moderate Risk category. The spread of points within each category also shows the variability of severity scores, indicating that High Risk drugs tend to exhibit both higher intensity and wider variation in side effects.
